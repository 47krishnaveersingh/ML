```
from urllib import request #to import from url
from bs4 import BeautifulSoup
import re #importing regular Expression
import time
#pointing to url
url='https://en.wikipedia.org/wiki/Machine_learning'

htmldata=request.urlopen(url) #downloading url data in html format
#htmldata.read() #It will show data
soupdata=BeautifulSoup(htmldata,'html5lib')
#html data,html parser--collection of html 5 data
#now selecting a particular tag for data scrape
ptagdata=soupdata.findAll('p')
# print(ptagdata)
#now converting data into string fotmat from html format
mydata=''
for i in ptagdata:
  mydata+=i.text
#print(mydata)



```

```
#Data Cleaning
clean_data=re.sub(r'\[[0-9]*\]',' ',mydata) #This will remove 0 or more times data appearing in mydata
clean_data=re.sub(r'\s+',' ',clean_data) #It will remove 1 or more whitespaces with single whitespace
clean_data=re.sub(r'\[^a-zA-Z]',' ',clean_data) #It will remove single char from starting of the line
clean_data=re.sub(r'\s+',' ',clean_data) #It will remove 1 or more whitespaces with single whitespace
print(clean_data)
print(type(clean_data))
#now Dealing With String
newdata=clean_data.split() #split converts string to list
# deal with data by using --nltk
import nltk
#Downloading all Dictioinary for computer
#nltk.download('all')
#removing Stopwords
mypuredata=[]
from nltk.corpus import stopwords
for i in newdata:
  if i.lower() not in stopwords.words('english'):
    mypuredata.append(i)

freq_data=nltk.FreqDist(mypuredata)
freq_data.plot(20)#will plot 20 most occuring words


```
