```
#importing matplotlib
import matplotlib.pyplot as plt
```

```
x=[2,9]
y=[7,15]

plt.xlabel('distance') #name of x axis
plt.ylabel('time') #name of y axis
plt.plot(x,y,label='Railway Track')     #this will plot line
plt.scatter(x,y,s=200,label='Railway Station',marker='*',color='red')    #this will plot dots
plt.bar(x,y,label='height of station') #this will plot bar graph
plt.grid(color='green')    #to draw grids
plt.legend()     #plot the labels
plt.show() #shows the graph
```

```
#this is data visualistion of cricket team 
players=['Dhoni','virat','Rohit','s jad']
scores=[120,140,180,89]

plt.xlabel('Players Name')
plt.ylabel('scores')
# plt.bar(players,scores,label='2020 cricket Match score IPL')
plt.pie(scores,labels=players,explode=(0.2,0,0,0),autopct='%1.1ff%%')
plt.grid(color='green')
plt.legend()
plt.show()
```
